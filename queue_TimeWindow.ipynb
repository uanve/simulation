```python
import simpy
import numpy as np
import itertools

class Queue():
    def __init__(self, env, N_m, N_e, N_n):
        
        self.Arrival = {}
        self.StartService = {}
        self.FinishService = {}
        
        self.Morning = simpy.Resource(env, N_m)
        self.Evening = simpy.Resource(env, N_e)
        self.Night   = simpy.Resource(env, N_n)
        self.env = env
                
        self.startMorning = 0
        self.startEvening = 30
        self.startNight = 60
    
    #arrivals/second
    def t_arrival(self,t):
        if t<self.startEvening:
            return 1
        elif t<self.startNight:
            return 1
        else:
            return 1

    def t_service(self):
        return 5

    def service(self,i):
        arrival_time = self.env.now
        
        # Add Arrival
        system.Arrival[i] = arrival_time
        
        


        
        if arrival_time==30:
            yield self.env.timeout(0.0001)
        elif arrival_time==60:
            yield self.env.timeout(0.0001)
        elif arrival_time==90:
            yield self.env.timeout(0.0001)

        # Morning shift 0-30
        if self.env.now < self.startEvening:
            
            with self.Morning.request() as req1:
                yield req1 | self.env.timeout(self.startEvening)
                
                if self.env.now < self.startEvening:
                    
                    system.StartService[i] = self.env.now
                    yield self.env.timeout(self.t_service())
                    
                    print("{} arrived at {} done at {} by 1".format(i, self.Arrival[i], self.env.now))
#                     print("{} arrived at {:.2f} done at {:.2f} by 1".format(i, self.Arrival[i], self.env.now))
                    self.FinishService[i] = self.env.now
                    
        
        # Evening shift 30-60
        if (self.startEvening <= self.env.now < self.startNight) & (i not in self.FinishService):
            
            with self.Evening.request() as req2:
                yield req2 | self.env.timeout(self.startNight)
                
                if self.env.now < self.startNight:
                    
                    system.StartService[i] = self.env.now
                    yield self.env.timeout(self.t_service())
                    
                    print("{} arrived at {} done at {} by 2".format(i, self.Arrival[i], self.env.now))
#                     print("{} arrived at {:.2f} done at {:.2f} by 1".format(i, self.Arrival[i], self.env.now))
                    self.FinishService[i] = self.env.now
            

        
#         # Night shift 60-90
        if (self.startNight <= self.env.now < 90) & (i not in self.FinishService):
            
            with self.Night.request() as req3:
                yield req3 | self.env.timeout(90)
                
                if self.env.now < 90:
                    
                    system.StartService[i] = self.env.now
                    yield self.env.timeout(self.t_service())
                    
                    print("{} arrived at {} done at {} by 3".format(i, self.Arrival[i], self.env.now))
#                     print("{} arrived at {:.2f} done at {:.2f} by 1".format(i, self.Arrival[i], self.env.now))
                    self.FinishService[i] = self.env.now
                else:
                    print("cannot finish {}".format(i))
        
    def arrivals(self):
        while self.env.now < 90:
            for i in itertools.count():
                self.env.process(self.service(i))
                t = self.t_arrival(self.env.now)
                yield self.env.timeout(t)  


```
